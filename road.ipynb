```python
from ultralytics import YOLO
import cv2
import yt_dlp


# STEP 1: Get YouTube stream URL
youtube_url = "https://www.youtube.com/watch?v=MncLrf2LsT8"

ydl_opts = {
    'format': 'best[ext=mp4]',  # get best mp4 stream
    'quiet': True
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(youtube_url, download=False)
    stream_url = info['url']

print("Stream URL fetched!")


# STEP 2: Load YOLOv8 model
model = YOLO('best-2.pt') # Loaded already pretrained model on Bdd100k dataset


# STEP 3: Open video stream
cap = cv2.VideoCapture(stream_url)

if not cap.isOpened():
    print("Error opening stream")
    exit()


# STEP 4: Tracking loop
while True:
    ret, frame = cap.read()
    if not ret:
        print("Stream ended or error")
        break

    # YOLOv8 tracking (your classes unchanged)
    results = model.track(
        frame,
        tracker="botsort.yaml",
        persist=True,
        conf=0.4
    )

    # Draw detections
    annotated_frame = results[0].plot()

    # Show output
    cv2.imshow("YOLOv8 Tracking", annotated_frame)

    # Exit key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


# CLEANUP
cap.release()
cv2.destroyAllWindows()
```
